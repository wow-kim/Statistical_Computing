### 부동소수점과 오차

```python
import math
from matplotlib import pyplot as plt
```

##### 부동 소수점은 컴퓨터에서 2진 소수로 표현됩니다.
##### 우리가 실생활에서 사용하는 대부분의 십진 소수는 정확하게 이진 소수료 표현될 수 없습니다.
##### 따라서, float자료형을 다룰 때는 항상 부동소수점 오차를 의식해야 합니다.
##### 예를 들면,

```python
-0.3 == -0.2 - 0.1
```

#### 해결책1 : `decimal.Decimal()`
##### 십진 소수 연산을 수행합니다.
#### 해결책2 : `math.isclose()`
#####

```python
import decimal
print( decimal.Decimal("-0.3") == decimal.Decimal("-0.2") + decimal.Decimal("-0.1") )

import math
print( math.isclose(-0.3, -0.1-0.2) )
```

- $ f(x) = {{1-\cos x} \over {x^2}} $
- ##### 위 방정식은 0 근방에서 부동소수점 연산에 의해 오차 발생 

```python
def evaluatefunction1(xmin, xmax,n):
    x = []
    f = []
    for i in range(n+1):
        try:
            x_val = xmin + i*(xmax-xmin)/n
            y_val = (1-math.cos(x_val)) / (x_val**2)
            x.append(x_val)
            f.append(y_val)
        except ZeroDivisionError:
            continue
    plt.plot(x,f)
```

```python
evaluatefunction1(-10**(-7),10**(-7),200)
```

- $ f(x) = {{\sin ^2 ({x \over 2}) } \over {(x^2) \over 2}} $
- ##### 동치인 함수, 0 근처에서 오차의 영향을 줄일 수 있다.

```python
def evaluatefunction2(xmin, xmax, n):
    x = []
    f = []
    for i in range(n+1):
        try:
            x_val = xmin + i*(xmax-xmin)/n
            y_val = math.sin(x_val/2)**2 / ( (x_val**2)/2 )
            x.append(x_val)
            f.append(y_val)
        except ZeroDivisionError:
            continue
    plt.plot(x,f)
```

```python
evaluatefunction2(-10**(-7),10**(-7),200)
```
